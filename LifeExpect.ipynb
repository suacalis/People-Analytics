```python
import plotly.express as px
```

```python
df =px.data.gapminder().query("year == 2007")
```

```python
df
```

```python
px.histogram(df, x="lifeExp",y="pop",marginal = "rug", hover_name="country",color="continent",height =700)
```

```python
px.bar(df, x="continent",y="pop", hover_name="country",color="lifeExp",height =700)
```

```python
px.sunburst(df,path=["continent","country"],values="pop",color="lifeExp",height=700)
```

```python
px.treemap(df,path=["continent","country"],values="pop",color="lifeExp",height=700)
```

```python
px.choropleth(df,locations="iso_alpha",hover_name="country",hover_data=df.columns,color="lifeExp",projection="natural earth",height=700)
```

```python
df =px.data.gapminder()
```

```python
px.scatter(df,
           x="gdpPercap",
           y="lifeExp",
           size="pop",
           size_max=65,
           hover_name="country",
           hover_data=df.columns,
           color="continent", 
           animation_frame="year",
           animation_group="country",
           range_y=[20,100],
           log_x=True,
           height=700,
          template="simple_white",
          title="Classic Gamminder Chart",
          labels=dict(gdpPercap="GDP per capita",
                     lifeExp="Life expectancy in years")).update_xaxes(tickprefix="$")
```
