```python
dfa[["datetime"]] = dfa[["Created"]].apply(pd.to_datetime)
#collect weekdays from datetime
dfa['day_of_week'] = dfa['datetime'].dt.day_name()
#collect hours from datetime
dfa['hour'] = dfa['datetime'].dt.strftime('%H').astype(int)
```

```python
ax = sns.countplot(x="day_of_week", data=dfa).set(title = 'Activity on days');
```

```python
ax = sns.countplot(x="hour", data=dfa).set(title = 'Activity by hour');
```
